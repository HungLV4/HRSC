```python
from BOVW import *
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = SVC(kernel="rbf", C=2.8, gamma=0.0073, cache_size=10000, probability=False)
# clf = RandomForestClassifier(n_estimators=61, max_depth=30, n_jobs=-1, random_state=111)
```

```python
feat = cv2.xfeatures2d.SIFT_create(100, edgeThreshold=50, contrastThreshold=0.02)
```

```python
bovw = BOVW(clf,
            feat,
            n_bags=100,
            tol=0.1,
            is_resample=True,
            is_reuse=False,
            verbose=True)
```

```python
%time bovw.fit("images/train/")
```

```python
%time acc = bovw.score("images/test/")
```
